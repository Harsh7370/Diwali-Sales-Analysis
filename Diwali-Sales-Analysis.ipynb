```python
sales = pd.read_csv('sales_data.csv',encoding = "Latin")
sales
```

```python
sales.shape
```

```python
sales.info()
```

```python
sales.columns
```

```python
sales.drop(['Status','unnamed1'],axis=1,inplace = True)
```

```python
sales.info()
```

```python
sales.dropna(inplace = True)
```

```python
sales
```

```python
sales.rename(columns={'Marital_Status':'Marital'},inplace =  True)
sales
```

```python
sales['Marital'].unique()
```

```python
sales[sales['Amount'].isnull()].count()
```

```python
sales['Marital']=sales['Marital'].map({0:'Single',1:'Married'})
sales['Gender']=sales['Gender'].map({'M':'Male','F':'Female'})
sales
```

```python
#Visualize the count of male and female based on gender column using seaborn
a = sns.countplot(x='Gender',data=sales,hue='Gender')
for i in a.containers:
    a.bar_label(i)
```

```python
#total spend by gender
a = sales.groupby('Gender')['Amount'].sum()
a=pd.DataFrame(a)
```

```python
b = sns.barplot(x='Gender',y='Amount',data=sales,hue='Gender',errorbar=None)
for i in b.containers:
    b.bar_label(i)
```
